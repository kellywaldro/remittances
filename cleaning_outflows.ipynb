# Cleaning and Merging

## Looking at money flows **from** countries

```python
import pandas as pd
```

```python
remittances = pd.read_csv('./wb_data/remittances/ouward_remittance_flows_brief_37_as_of_nov.28_2022.csv')
```

```python
remittances.columns
```

```python
remittances.head()
```

```python
remittances = remittances[['Country','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2021 Nominal GDP ']]
```

```python
remittances.rename(columns={'Country':'country'}, inplace=True)
```

```python
income_levels = pd.read_csv('./countries_by_income.csv')
```

```python
remittances = pd.merge(remittances, income_levels, on='country', how='outer')
```

```python
remittances_gdp = remittances[['country','income_level','2021 Nominal GDP ']]
```

```python
remittances = remittances.drop(columns=['2021 Nominal GDP '])
```

```python
# drop null values

remittances = remittances.dropna()
```

```python
remittances
```

```python
migrant_stock = pd.read_csv('./wb_data/migrant_stock/API_SM.POP.TOTL.ZS_DS2_en_csv_v2_4772109.csv')
```

```python
migrant_stock.rename(columns={'Country Name':'country'}, inplace=True)
```

```python
migrant_stock = migrant_stock[['country', 'Indicator Name','2015']]
```

```python
remittances = pd.merge(remittances, migrant_stock, on='country', how='inner')
```

```python
remittances = remittances[['country','income_level', '2021','2015_y']]
```

```python
remittances = remittances.dropna()
```

```python
# use altair to create a scatter plot with the remittances and migrant stock data

import altair as alt

alt.Chart(remittances).mark_circle().encode(
    x='2021',
    y='2015_y',
    color='income_level',
    tooltip=['country', 'income_level', '2021','2015_y']
).interactive()
```

```python
remittances['2021'] = remittances['2021'].str.replace(',', '')
```

```python
# save remittances to csv

remittances.to_csv('./outward_remittances.csv', index=False)
```
