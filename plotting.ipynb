# Plotting the Data

```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('./fdi_remit_oda.csv')
```

```python
df
```

```python
df['remittances_total'] = df['remittances_total']/1000000000

df['oda_total'] = df['oda_total']/1000000000

df['fdi_total'] = df['fdi_total']/1000000000
```

```python
df
```

```python
# remove the data for 2021 - because we don't have the ODA for that year 

df = df[df['year'] != 2021]
```

```python
df_melt = pd.melt(df, id_vars='year', value_vars=['fdi_total', 'remittances_total', 'oda_total'], var_name='type', value_name='value')
```

```python
df_melt.to_clipboard(index=False)
```

```python
df
```

```python
base = alt.Chart(df.reset_index()).encode(x='year')

alt.layer(
    base.mark_line(color='blue').encode(y='oda_total'),
    base.mark_line(color='red').encode(y='remittances_total')
).properties(
    width=650,
    height=450
)
```

```python
base = alt.Chart(df.reset_index()).encode(x='year')

alt.layer(
    base.mark_line(color='blue').encode(y='oda_total'),
    base.mark_line(color='red').encode(y='remittances_total'),
    base.mark_line(color='green').encode(y='fdi_total'),
).properties(
    width=650,
    height=450
)
```

```python
df.to_clipboard(index=False)
```
