# Country classification by income level

Taking the World Bank's country classification by income level (as of 2023) and compiling it into one dataset

```python
import pandas as pd
```

```python

tables = pd.read_html("https://datahelpdesk.worldbank.org/knowledgebase/articles/906519-world-bank-country-and-lending-groups")
```

```python
low_df = tables[8]
low_middle_df = tables[9]
upper_middle_df = tables[10]
high_df = tables[11]
```

```python
# take every cell in the table and put it in a list

low_list = low_df.values.tolist()
low_middle_list = low_middle_df.values.tolist()
upper_middle_list = upper_middle_df.values.tolist()
high_list = high_df.values.tolist()
```

```python
low_list = [x for x in low_list if str(x) != 'nan']
low_middle_list = [x for x in low_middle_list if str(x) != 'nan']
upper_middle_list = [x for x in upper_middle_list if str(x) != 'nan']
high_list = [x for x in high_list if str(x) != 'nan']
```

```python
low_income_countries = []

for x in low_list:
    low_income_countries.append(x[0])
    low_income_countries.append(x[1])
    low_income_countries.append(x[2])
```

```python
low_middle_countries = []

for x in low_middle_list:
    low_middle_countries.append(x[0])
    low_middle_countries.append(x[1])
    low_middle_countries.append(x[2])
```

```python
upper_middle_countries = []

for x in upper_middle_list:
    upper_middle_countries.append(x[0])
    upper_middle_countries.append(x[1])
    upper_middle_countries.append(x[2])
```

```python
high_countries = []

for x in high_list:
    high_countries.append(x[0])
    high_countries.append(x[1])
    high_countries.append(x[2])
    
```

```python
income_dict = {}

for x in low_income_countries:
    income_dict[x] = "low income"

for x in low_middle_countries:
    income_dict[x] = "low middle income"

for x in upper_middle_countries:
    income_dict[x] = "upper middle income"

for x in high_countries:
    income_dict[x] = "high income"
    
income_dict
```

```python
import csv

with open('countries_by_income.csv', 'w') as csv_file:
    writer = csv.writer(csv_file)
    for key, value in income_dict.items():
       writer.writerow([key, value])
```
