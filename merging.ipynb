# Cleaning and Merging

## Taking the World Bank's remittance and FDI data and putting them together

```python
import pandas as pd
```

```python
fdi_df =pd.read_csv('./FDI_inflows/API_BX.KLT.DINV.CD.WD_DS2_en_csv_v2_4770445.csv')

fdi_df.rename(columns={'Country Name':'country'},inplace=True)

fdi_df
```

```python
# read csv and make a header

income_df = pd.read_csv('./countries_by_income.csv')
```

```python
income_df['country'] = income_df['country'].str.replace('"', '')
income_df['country'] = income_df['country'].str.replace('ü', 'u')
income_df['country'] = income_df['country'].str.replace('Czech  Republic', 'Czechia')
income_df['country'] = income_df['country'].str.replace('ô', 'o')
income_df['country'] = income_df['country'].str.replace('Iran, Islamic Rep', 'Iran, Islamic Rep.')
income_df['country'] = income_df['country'].str.replace('São Tomé and Principe', 'Sao Tome and Principe')
income_df['country'] = income_df['country'].str.replace('Congo, Dem. Rep', 'Congo, Dem. Rep.')
income_df['country'] = income_df['country'].str.replace('Curaçao','Curacao')
income_df['country'] = income_df['country'].str.replace('\ufeff\ufeffSyrian Arab Republic','Syrian Arab Republic')
income_df['country'] = income_df['country'].str.replace("Korea, Dem. People's Rep", "Korea, Dem. People's Rep.")
income_df['country'] = income_df['country'].str.replace('  ', ' ')
```

```python
# merge the df using an outer join

merged_df = pd.merge(fdi_df, income_df, on='country', how='outer')
```

```python
merged_df
```

```python
merged_df = merged_df[:-2]
```

see world bank website about [data for Taiwan](https://datahelpdesk.worldbank.org/knowledgebase/articles/114933-where-are-your-data-on-taiwan#:~:text=Taiwan%2C%20China%2C%20is%20not%20listed,the%20high%2Dincome%20countries%20aggregate.)

```python
merged_df
```

```python
# list the columns

merged_df.columns
```

```python
# keep the year columns from 1990 to 2021 

merged_df = merged_df[['country','Country Code','income_level','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]

merged_df
```

```python
# group by income level to make a new df and sum the values 

grouped_df = merged_df.groupby('income_level').sum()
```

```python
grouped_df = grouped_df.transpose()
```

```python
# rename columns

grouped_df.rename(columns={'income_level':'year'},inplace=True)
```

```python
grouped_df
```

```python
grouped_df["fdi_total"] = grouped_df["low income"] + grouped_df["low middle income"] + grouped_df["upper middle income"]

# only keep the year and fdi_total columns

fdi_df = grouped_df[['fdi_total']]
```

```python

fdi_df.columns
```

```python
# rename the index to year

fdi_df.index.name = 'year'
```

```python
fdi_df.reset_index(inplace=True)
```

```python
fdi_df
```

```python
remittances = pd.read_csv('./inward_remittance_flows_brief_37_revised_final_version_as_of_dec._2_2022.csv')
```

```python
remittances.columns
```

```python
# rename columns

remittances.rename(columns={'Migrant remittance inflows (US$ million)':'remittances'},inplace=True)
```

```python
# keep rows where it is equal to a certain value

remittances = remittances[remittances['remittances'] == 'Low-and Middle-Income Countries']
```

```python
remittances
```

```python
# keep the columns from 1990 to 2021

remittances = remittances[['remittances','1990','1991','1992','1993','1994','1995','1996','1997','1998','1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021']]
```

```python
remittances = remittances.transpose()
```

```python
remittances
```

```python
# delete the first row

remittances = remittances.drop(remittances.index[0])
```

```python
# rename the columns

remittances.rename(columns={214:'remittance_millions'},inplace=True)
```

```python
remittances

# make index a column

remittances.reset_index(inplace=True)
```

```python
# rename the index column

remittances.rename(columns={'index':'year'},inplace=True)
```

```python
# merge remittances and fdi

fdi_remit_df = pd.merge(fdi_df, remittances, on='year', how='outer')

fdi_remit_df
```

```python
# divide the fdi column by 1000000

fdi_remit_df['remittance_millions'] = fdi_remit_df['remittance_millions'].str.replace(',', '')
fdi_remit_df['remittance_millions'] = fdi_remit_df['remittance_millions'].astype(float) * 1000000
```

```python
fdi_remit_df
```

```python
# save to csv

fdi_remit_df.to_csv('fdi_remit.csv', index=False)
```
